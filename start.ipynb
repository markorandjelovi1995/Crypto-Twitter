```python
import dotenv
import datetime as dt
import os
from pathlib import Path
from db import (
    get_daily_statuses,
    get_db_object,
    get_monthly_statuses,
    get_statuses_from_db,
    get_weekly_statuses,
)
from status import (
    get_top_statuses,
    statuses_with_score_to_str,
)
```

# Initialization

```python
dotenv.load_dotenv()
```

```python
db_path = os.getenv("DB_PATH")
db = get_db_object(Path(db_path))
```

# Commands

## Daily

```python
# top 5 statuses
# on 2021-07-08
# not retweet (can be tweet, reply or quote)
date = dt.date(2021, 7, 8)
tweet = None
reply = None
retweet = False
quote = None
count = 5
statuses = get_daily_statuses(db, date, tweet, reply, retweet, quote)
top_statuses_with_score = get_top_statuses(statuses, count)
result_str = statuses_with_score_to_str(top_statuses_with_score)

print(result_str)
```
